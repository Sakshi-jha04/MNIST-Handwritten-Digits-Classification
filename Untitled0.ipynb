```python
data=np.array(data)
m,n=data.shape

np.random.shuffle(data)
data_dev=data[0:1000].T
Y_dev=data_dev[0]
X_dev=data_dev[1:n]
X_dev=X_dev/255

data_train=data[1000:m].T
Y_train=data_train[0]
X_train=data_train[1:n]
X_train=X_train/255
m=Y_train.size
```

```python
def reLU(Z):
  return np.maximum(0,Z)

def softmax(Z):
  return np.exp(Z)/sum(np.exp(Z))

def deriv_reLU(Z):
  return Z > 0

def Y_cheat(Y):
  Y_cheat_new=np.zeros((Y.size,Y.max()+1))
  Y_cheat_new[np.arange(Y.size),Y]=1
  Y_cheat_new=Y_cheat_new.T
  return Y_cheat_new

def init_params():
  W1=np.random.randn(10,784)
  b1=np.random.randn(10,1)
  W2=np.random.randn(10,10)
  b2=np.random.randn(10,1)
  return W1,b1,W2,b2

def for_prop(W1, b1, W2, b2, X):
    Z1 = np.dot(W1, X) + b1
    A1 = reLU(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = softmax(Z2)
    return Z1, A1, Z2, A2

def back_prop(Z1,A1,Z2,A2,W2,X,Y):
  m=Y.size

  Y_cheat_new=Y_cheat(Y)
  dz2=A2-Y_cheat_new
  dw2=1/m*dz2.dot(A1.T)
  db2=1/m*np.sum(dz2)
  dz1=W2.T.dot(dz2)*deriv_reLU(Z1)

  dw1=1/m*dz1.dot(X.T)
  db1=1/m*np.sum(dz1)

  return dw1, db1, dw2, db2

def new_params(W1,b1,W2,b2,dw1,db1,dw2,db2,alpha):
  W1=W1-alpha*dw1
  b1=b1-alpha*db1

  W2=W2-alpha*dw2
  b2=b2-alpha*db2
  return W1,b1,W2,b2
```

```python
def get_pred(A2):
  return np.argmax(A2,0)

def get_acc(pred,Y):
  print(pred,Y);
  return np.sum(pred==Y)/Y.size

def gradient_decent(X,Y,epoch,alpha):
  W1,b1,W2,b2=init_params()
  for i in range(epoch):
    Z1,A1,Z2,A2=for_prop(W1,b1,W2,b2,X)
    dw1, db1, dw2, db2=back_prop(Z1,A1,Z2,A2,W2,X,Y)
    W1,b1,W2,b2=new_params(W1,b1,W2,b2,dw1,db1,dw2,db2,alpha)
    if (i%50==0):
      print('epoch : ', i)
      print('accuracy : ',get_acc(get_pred(A2),Y))
  return W1,b1,W2,b2
```

```python
W1,b1,W2,b2=gradient_decent(X_train,Y_train,2000,0.3)
```

```python
def make_predictions(X, W1, b1, W2, b2):
    _, _, _, A2 = for_prop(W1, b1, W2, b2, X)
    predictions = get_pred(A2)
    return predictions

def test_prediction(index, W1, b1, W2, b2):
    current_image = X_train[:, index, None]
    prediction = make_predictions(X_train[:, index, None], W1, b1, W2, b2)
    label = Y_train[index]
    print("Prediction: ", prediction)
    print("Label: ", label)

    current_image = current_image.reshape((28, 28)) * 255
    plt.gray()
    plt.imshow(current_image, interpolation='nearest')
    plt.show()
```

```python
test_prediction(0, W1, b1, W2, b2)
test_prediction(1, W1, b1, W2, b2)
test_prediction(2, W1, b1, W2, b2)
test_prediction(3, W1, b1, W2, b2)
```
